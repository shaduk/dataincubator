```r
agedata <- select(mydata, sex, age, snsjob1a)
```

```r
agedata <- agedata[agedata$age != 98 & agedata$age != 99,]
```

What is the mean age of males and females who participated in the survey?

```r
agedata %>% group_by(sex) %>% summarise(mean(age))
```

Hence the mean of the age of males is 49.12 and mean of the age of females is 52.94

What is the variation between sex, age and the usage of social media sites to look for or research a job?

```r
qplot(sex, age,data=agedata,colour = snsjob1a)
```

This shows the variation in males and females who have used social media sites to look for a job

```r
agedata1 <- select(mydata, sex, age, cable3a)
```

```r
agedata1 <- na.omit(agedata1)
```

```r
agedata1 <- agedata1[agedata1$age!=98 & agedata1$age!=99,]
```

```r
agedata1 <- agedata1[agedata1$cable3a != 8 & agedata1$cable3a != 9 & agedata1$cable3a != 2,]
```

Which age segment is more likely to use the dating app on phone ?
